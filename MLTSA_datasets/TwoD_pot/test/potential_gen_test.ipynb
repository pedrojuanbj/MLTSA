```python
def spiral_potential_3b(x, y):
    return -(0.05*x**2 + 0.05*y**2 + 1)*np.exp(-1.02040816326531/(0.05*(x**2 + y**2)*np.sin(0.03*x**2 + 0.03*y**2 + 3*np.arctan2(x, y)) + 1)**2)
# Meshgrid of potential surface
[x,y] = np.meshgrid(np.linspace(-20, 20, 200),np.linspace(-20, 20, 200))

# Calculatin Energy for each
z = spiral_potential_3b(x, y)

#Plotting the potential Surface
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, cmap=plt.cm.viridis, alpha=0.6)
```
